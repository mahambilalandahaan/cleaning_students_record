```python
#cleaning plot by adding a column score fill missing place of the row with medain
#compare bar diagrame and after
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
student={"Name":["Atif","Maham","Bilal","atif","ATIF","Madiha","maham"],
         "Age":[46,15,13,46,46,None,None],
         "Class":["MS","Inter","8th","MS","MS",None,"iNTER"]}
df=pd.DataFrame(student)
missing_before=df.isnull().sum()
lower_case=df["Name"].str.lower().str.strip()
lower_case1=df["Class"].str.lower().str.strip()
df["Name"]=lower_case
df["Class"]=lower_case1
df["Name"].duplicated()
remove_duplicate=df.drop_duplicates(subset="Name",keep="first")
df=remove_duplicate
fill=df["Age"].fillna(df["Age"].mean())
df["Age"]=fill
df["Class"]=df["Class"].fillna("unknown")
df["Score"]=[90,85,None,70]
df["Score"]=df["Score"].fillna(df["Score"].median())
print("FINAL CLEANED DATAFRAME \n",df)
df_cleaned=df
missing_after = df_cleaned.isnull().sum()

pd.DataFrame({"Before": missing_before, "After": missing_after}).plot(
    kind="bar", color=["orange", "green"]
)
plt.title("Missing Values Before vs After Cleaning")
plt.show()

```
